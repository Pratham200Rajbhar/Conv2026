```python
from tensorflow.keras.models import load_model

CROP_NAME = "Brinjal"

model_path = f"/content/drive/MyDrive/Ganpat_hackathon_2026/ML1/Train_model/{CROP_NAME}_model.keras"
model = load_model(model_path)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

test_dir = f"/content/drive/MyDrive/Ganpat_hackathon_2026/ML1/split_dataset/test/{CROP_NAME}"

test_datagen = ImageDataGenerator(rescale=1./255)

test_data = test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=16,
    class_mode='categorical',
    shuffle=False
)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

test_dir = f"/content/drive/MyDrive/Ganpat_hackathon_2026/ML1/split_dataset/test/{CROP_NAME}"

test_datagen = ImageDataGenerator(rescale=1./255)

test_data = test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=16,
    class_mode='categorical',
    shuffle=False
)
```

```python
import numpy as np

predictions = model.predict(test_data)
y_pred = np.argmax(predictions, axis=1)
y_true = test_data.classes
```

```python
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix

# Get class labels in order of their numerical index
class_labels = [None] * len(test_data.class_indices)
for name, idx in test_data.class_indices.items():
    class_labels[idx] = name

unique_labels = np.unique(y_true)
filtered_class_labels = [class_labels[i] for i in unique_labels]

print(classification_report(y_true, y_pred, labels=unique_labels, target_names=filtered_class_labels))
```

```python
cm = confusion_matrix(y_true, y_pred)
print(cm)
```

```python
loss, acc = model.evaluate(test_data)
print("Test Accuracy:", acc)
```
