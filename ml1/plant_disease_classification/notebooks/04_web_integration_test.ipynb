```python
models = {
    "Brinjal": load_model("/content/drive/MyDrive/Ganpat_hackathon_2026/ML1/Train_model/Brinjal_model.keras"),
    "Guava": load_model("/content/drive/MyDrive/Ganpat_hackathon_2026/ML1/Train_model/Guava_model.keras"),
    "Papaya": load_model("/content/drive/MyDrive/Ganpat_hackathon_2026/ML1/Train_model/Papaya_model.keras"),
    "Castor": load_model("/content/drive/MyDrive/Ganpat_hackathon_2026/ML1/Train_model/Castor_model.keras"),
    "Cumin": load_model("/content/drive/MyDrive/Ganpat_hackathon_2026/ML1/Train_model/Cumin_model.keras"),
}
```

```python
class_labels = {
    "Brinjal": ['Healthy', 'Bacrerial_Wilt', 'Mosaic', 'Cercospora_Leaf_spot', 'Phomopsis_Leaf_Blight'],
    "Castor": ['Healthy', 'leaf_curv_virus', 'Bacterial_Leaf_Blight', 'Alternaria_Leaf_Blight', 'Cercospora_Leaf_Spot'],
    "Papaya": ['Healthy', 'Leaf_Spot', 'Powdery_Mildew', 'Ring_Spot_Virus'],
    "Cumin": ['Healthy', 'Alternaria_Blight', 'Wilt'],
    "Guava": ['Healthy', 'Bacterial_Blight', 'Anthracnose', 'Wilt', 'Red_Rust']
}
```

```python
def preprocess_image(img_path):
    img = cv2.imread(img_path)
    img = cv2.resize(img, (224, 224))
    img = img / 255.0
    img = np.expand_dims(img, axis=0)
    return img
```

```python
def predict_image(img_path, crop):
    model = models[crop]
    labels = class_labels[crop]

    img = preprocess_image(img_path)
    pred = model.predict(img)

    class_index = np.argmax(pred)
    confidence = np.max(pred)

    return labels[class_index], confidence
```

```python
crop_name = "Castor"   # change manually for testing
```

```python
label, confidence = predict_image(img_path, crop_name)

print("Crop:", crop_name)
print("Prediction:", label)
print("Confidence:", round(confidence*100, 2), "%")
```
